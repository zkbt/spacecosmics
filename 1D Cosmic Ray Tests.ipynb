# 1D Cosmic Ray Tests

This notebook shows 1D tests of different cosmic ray mitigation techniques. It creates 1D timeseries, and then bins then according to different techniques. For these tests, we care about a couple of items:
+ How much extra noise to we introduce into binned light curves, just by using an algorithm that's throwing out some of the data?
+ How well does the algorithm eliminate or reduce cosmic ray hits?
To try to capture both of these points, we make plots showing the achieved standard deviation of the binned data for each different method, as a function of the ratio of the cosmic ray strength to the original binned noise. Most methods (except a straight `Mean`, of course) do a pretty good job of eliminating cosmic rays, but some with long timescale associated with them can fail to perfect zap every cosmic -- these show up as trends of increasing noise with increasing amplitude of cosmic ray strength.

(Note that while we talk about a "cosmic ray strength" in these simple tests, it's probably better to think of this in terms of brighter or fainter stars. Cosmic ray hits are about the same no matter what, but they would appear "weaker" for a bright star where they are a fractionally small ratio of the total flux and "stronger" for a faint star where they are a larger fraction of the total flux.)

```python
from comparemethods1D import *
%matplotlib inline
```

We can test an individual strategy and timeseries as follows:

```python
# create a cosmic ray mitigation strategy
strategy = central(10)

# create a timeseries to test it on
timeseries = Timeseries1D(nexposures=1324, nsubexposures=900)

# apply the strategy to the timeseries
strategy.calculate(timeseries)

# plot the result
strategy.plot()
```

We can use these plots to get a decent visual sense of how well the mitigation is doing. The blue is the light curve with cosmic rays injected. The orange is what the light curve would be if no cosmic rays had been inject at all. The purple shows what comes out of the mitigation algorithm -- ideally it should overlap the orange curve fairly closely. Note that the sideways histogram on the right side is on a log scale, so a Gaussian distribution should look like a perfect parabola. In this example, the unmitigated light curve shows a significant tail toward high values -- those are the cosmics!

To compare different methods and parameters for methds, the `compare` function in `comparemethods1D` loops over a suite of them. It can take multiple iterations (realizing new light curves each time), and therefore provide more precise estimates of the achieved noise possible with each one. Note that you might want to run with `niterations=1` first to check it out, and then ramp up to `niterations=100` for less noisy curves on the comparison plots.

```python
compare(nexposures=1324, # how many binned exposures? (e.g. 1324 FFIs in one sector)
        nsubexposures=900, # how many sub-exposures in each exposure (900 two-seconds in one 30 minute)
        niterations=100, # how many times should we do the test? (more gets prettier plots)
        remake=False) # if the data for this already exist, should we remake anyway?
```
